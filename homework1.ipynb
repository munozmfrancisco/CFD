## Exercise 2

Plot the function:

$$
f(x) = \frac{cos(x)}{x}, \qquad \text{where} \quad x \in \quad ] 0, \pi ]
$$



and the fourth order Taylor Series at the point $x = \frac{\pi}{2}$:

$$
f_T(x) = -\frac{2}{\pi}(x-\frac{\pi}{2}) + \frac{4}{\pi^2}(x-\frac{\pi}{2})^2 + (\frac{1}{3\pi} - \frac{8}{\pi^3})(x-\frac{\pi}{2})^3 + (-\frac{2}{3\pi^2} + \frac{16}{\pi^4})(x-\frac{\pi}{2})^4.
$$

First, we import the libraries:

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

Now, we define the $cosine$ function and set the value of $\pi$.

```python
cos = np.cos
pi = math.pi
```

In addition, we define the domain and the function $f(x)$.

```python
n = 101
x = np.linspace(0, pi, n)[1:] #Drop the first element of the interval.
f = cos(x)/x
```

Here, we introduce the Taylor polynomial function.

```python
def f_T(x):
    term = x - pi / 2
    f1 = -2 / pi
    f2 = 8 / pi**2
    f3 = 2 / pi - 48 / pi**3
    f4 = -16 / pi**2 + 384 / pi**4
    return f1 * term  + f2 * term**2 / 2 + f3 * term**3 / 6 + f4 * term**4 / 24
```

Finally, we plot the function $f(x)$ and its corresponding Taylor series $f_T(x)$. 

```python
plt.figure(figsize=(7,5))
plt.plot(x, f, '.-', label = r'$f(x)$')
plt.plot(x, f_T(x), '-', label = r'$f_T(x)$')
plt.xlabel('$x$')
plt.title('$f(x)$ and $f_T(x)$ Plot')
plt.legend()
```

## Exercise 3

For the functions:

$$
F_1 (x,y) = \sqrt{x^2 + y ^2}, \\
F_2 (x,y) = t(x + y), \\
F_3 (x,y) = sin(2 \pi \sqrt{x^2 + y^2}), 
$$

where $(x, y) \in [-3,3]^2$,

make a contour plots and 3D surface plots.

First, we define the functions and the domain.

```python
x = np.linspace(-3, 3, 50)
y = np.linspace(-3, 3, 50)

t = 1.
sin = np.sin
sqrt = np.sqrt

X, Y = np.meshgrid(x, y)

F_1 = sqrt(X**2 + Y**2)
F_2 = t * (X + Y)
F_3 = sin(2 * pi * F_1)
```

Finally, we make the contour plots of the functions:

```python
plt.figure()
pcm = plt.contourf(X, Y, F_1, cmap='viridis')
plt.colorbar(pcm, label=r'$F_1(x,y)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$F_1(x,y)$ Contour Plot')


plt.figure()
pcm = plt.contourf(X, Y, F_2, cmap='viridis')
plt.colorbar(pcm, label=r'$F_2(x,y)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$F_2(x,y)$ Contour Plot')


plt.figure()
pcm = plt.contourf(X, Y, F_3, cmap='viridis')
plt.colorbar(pcm, label=r'$F_3(x,y)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$F_3(x,y)$ Contour Plot')
```

and plot the 3D surfaces.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, F_1, cmap='viridis', linewidth=0, antialiased=True)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title('$F_1(x,y)$ 3D Surface')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, F_2, cmap='viridis', linewidth=0, antialiased=True)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title('$F_2(x,y)$ 3D Surface')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, F_3, cmap='viridis', linewidth=0, antialiased=True)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title('$F_3(x,y)$ 3D Surface')
```

## Exercise 5

Considering this curves:

$$
\overrightarrow{c_0}(t) := \overrightarrow{f}(t; x_0 = 0, y_0 = 0),
$$

$$
\overrightarrow{c_1}(t) := \overrightarrow{f}(t; x_0 = 0.5, y_0 = 0.5),
$$

$$
\overrightarrow{c_2}(t) := \overrightarrow{f}(t; x_0 = 1, y_0 = 1).
$$

where $t \in [0,6\pi]$




First, we define the domain and the curves.

```python
t = np.linspace(0, 6*pi, 101)

c_0 = (0 + sin(t), 0 + sin(t))
c_1 = (0.5 + sin(t), 0.5 + sin(t))
c_2 = (1 + sin(t), 1 + sin(t))
```

Finally, we plot the curves, setting the z-axis as t.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(*c_0, t, color='red', lw=2)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$t$')
ax.set_title('$\overrightarrow{c_0}$ 3D Plot')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(*c_1, t, color='blue', lw=2)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$t$')
ax.set_title('$\overrightarrow{c_1}$ 3D Plot')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(*c_2, t, color='green', lw=2)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$t$')
ax.set_title('$\overrightarrow{c_2}$ 3D Plot')
```
